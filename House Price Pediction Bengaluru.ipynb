```python
data['location'] = data['location'].apply(lambda x : x.strip())
```

```python
location_stats = data.groupby('location')['location'].agg('count').sort_values(ascending = False)
```

```python
location_stats
```

```python
location_less_than_10_entries = location_stats[location_stats<=10]
```

```python
location_less_than_10_entries.sum()
```

```python
data['location'] = data['location'].apply(lambda x : 'others' if x in location_less_than_10_entries else x)
```

```python
data['location'].value_counts()
```

```python
data
```

```python
data['size'].unique()
```

```python
data['Bedroom'] = data['size'].apply(lambda x:int ( x.split(' ')[0]))
```

```python
data
```

```python
data.total_sqft.unique()
```

```python
def clean(sqft): 
    tokens = sqft.split('-')
    if len(tokens) == 2 :
        return (float(tokens[0]) + float(tokens[1]))/2 
    else: 
        try:
            return float(sqft)
        except: 
            return None  

clean('total_sqft')
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data['sqft_per_bed'] = data['total_sqft']/ data['Bedroom']
```

```python
data
```

```python
data.sqft_per_bed.describe()
```

```python
data2 = data[data['sqft_per_bed'] >= 300]
```

```python
data2.loc[:, 'price_per_sqft'] = round( data2['price']*100000/data2['total_sqft'], 2)
```
